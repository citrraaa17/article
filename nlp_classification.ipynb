```python
import numpy as np
import pandas as pd
import tensorflow as tf
import re
import nltk
nltk.download('stopwords')
```

```python
dftr=pd.read_csv('/content/train.csv')
dftr
```

```python
dftr.isna().sum()
```

```python
dftr=dftr.drop_duplicates()
```

```python
dftr['Class Index'].value_counts()
```

```python
def preprocess_text(text):
    # Ubah huruf besar ke huruf kecil
    text = text.lower()

    # Hapus simbol dan angka kecuali huruf
    text = re.sub(r"[^a-zA-Z\s]", '', text)

    # Hapus spasi berlebih
    text = re.sub(r"\s+", ' ', text).strip()

    # Hapus stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    text = ' '.join([word for word in words if word not in stop_words])

    return text


dftr['Title'] = dftr['Title'].apply(preprocess_text)
dftr['Description'] = dftr['Description'].apply(preprocess_text)
dftr
```

```python
dft=pd.read_csv('/content/test.csv')
```

```python
dft.isna().sum()
```

```python
dft['Title'] = dft['Title'].apply(preprocess_text)
dft['Description'] = dft['Description'].apply(preprocess_text)
dft
```

```python
X_train=dftr['Description']
y_train=dftr['Class Index']
```

```python
X_test=dft['Description']
y_test=dft['Class Index']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
print(f"X_train_tfidf shape: {X_train_tfidf.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test_tfidf shape: {X_test_tfidf.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import to_categorical

num_classes = 5
input_shape = (84536,)

# Membangun model
model = Sequential([
    Input(shape=input_shape),  # Menggunakan Input layer dengan input shape yang benar
    Dense(641, activation='relu'),
    Dense(32, activation='relu'),
    Dense(num_classes, activation='softmax')  # Softmax untuk multiclass classification
])

# Kompilasi model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Loss untuk multiclass
              metrics=['accuracy'])

# Melatih model dengan data yang benar
history = model.fit(X_train_tfidf, y_train,  # Gunakan X_train_tfidf
                    epochs=20,
                    batch_size=64,
                    validation_data=(X_test_tfidf, y_test))  # Gunakan X_test_tfidf

# Evaluasi model
test_loss, test_accuracy = model.evaluate(X_test_tfidf, y_test)  # Gunakan X_test_tfidf
print(f"Test accuracy: {test_accuracy}")
```
